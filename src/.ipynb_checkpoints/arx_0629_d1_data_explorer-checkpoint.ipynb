### LOAD DATA ###

```python
#LOAD DATA FILES
from _config import OLIGO_LIBRARY, POSITIVE_CONTROLS_FILE
import numpy as np
import pandas as pd, re

#LOAD DATA
oligo_exp_info= pd.read_csv("../out/0628_d1a_all_exp_oligo_info.csv")
oligos = pd.read_csv("../out/0628_d1a_all_oligo_info.csv")
positive_controls = pd.read_csv(POSITIVE_CONTROLS_FILE)

#DEFINE CONSTANTS
MODERATE_EXPR_MU = 15
HIGH_EXPR_MU = 50
```

### ANNOTATE OLIGOS ###

```python
oligos_lib = OLIGO_LIBRARY
oligos_lib["id"] = oligos_lib.index
oligos_lib["starts"] = oligos_lib.chromosome_info.apply(lambda x: int(re.compile("(\d+)-(\d+)").search(x).groups()[0]))
oligos_lib["starts"] = oligos_lib.starts-oligos_lib.starts.min()

candid_regex = re.compile("candid_(\d+)")
neg_regex = re.compile("Neg_(\d+)")
mutant_regex = re.compile("mut(\d+)$")

oligos_lib["locus_ids"] = oligos_lib.RefSeqID.apply(lambda x: int(candid_regex.search(x).groups()[0]) if candid_regex.search(x) else neg_regex.search(x).groups()[0] )
oligos_lib["mutant_num"] = oligos_lib.RefSeqID.apply(lambda x: int(mutant_regex.search(x).groups()[0]) if mutant_regex.search(x) else 0 )
oligos_lib["is_cand"] = oligos_lib.RefSeqID.apply(lambda x: True if candid_regex.search(x) else False )
oligos_lib["is_neg"] = oligos_lib.RefSeqID.apply(lambda x: True if neg_regex.search(x) else False )
oligos_lib["mutant_start"] = oligos_lib.starts + oligos_lib.mutant_num*30 - 30


unique_starts = pd.DataFrame(pd.Series(oligos_lib.starts.unique()).rename("starts"))
start_indexes = unique_starts.reset_index().set_index("starts")
oligos_lib["start_index"] = oligos_lib.starts.apply(lambda x: start_indexes.loc[x])

oligos = oligos.set_index("oligo")
oligos = oligos.join(oligos_lib, on ="oligo")#.dropna()
oligos["mu"] = oligos.n_transcripts / oligos.n_bcs

positive_control_oligos =  oligos.loc[oligos.RefSeqID.isin(positive_controls.Name)]
positive_control_positions = positive_control_oligos.starts
wt_controls = oligos.loc[(oligos.mutant_num==0) &(oligos.starts.isin(positive_control_positions))]
mutated_controls = oligos.loc[(oligos.mutant_num>0)&(oligos.starts.isin(positive_control_positions))]

oligos_by_exp = oligo_exp_info
oligos_by_exp = oligos_by_exp.join(oligos_lib, on ="oligo")#.dropna()
oligos_by_exp["mu"] = oligos_by_exp.n_transcripts / oligos_by_exp.n_bcs
mean_mu = oligos_by_exp.mu.mean()
mu_by_exp = oligos_by_exp.groupby("exp").mu.mean()
oligos_by_exp["exp_mu"] =  oligos_by_exp.exp.apply(lambda x: mu_by_exp.loc[x])
oligos_by_exp["exp_norm_mu"]= (oligos_by_exp["mu"] / oligos_by_exp["exp_mu"]) * mean_mu
oligos_by_exp["exp_nm"] = oligos_by_exp.apply(lambda x:re.compile("_BR.").sub("",x.exp),axis=1)          
oligos_by_exp["rep"] = oligos_by_exp.apply(lambda x:re.compile("_BR(.)").search(x.exp).group(1) if "BR" in x.exp else None,axis=1)
oligos_by_exp= oligos_by_exp.loc[oligos_by_exp.exp != "other"]

```

### IDENTIFY ENHANCERS  ###

```python
#categorize oligo groups by mean expression
starts_by_mean_expression = oligos.groupby("start_index").mu.mean().sort_values(ascending=False)
low_transcribed_starts = starts_by_mean_expression.loc[(starts_by_mean_expression<MODERATE_EXPR_MU)]

moderately_transcribed_starts = starts_by_mean_expression.loc[(starts_by_mean_expression>=MODERATE_EXPR_MU)&(starts_by_mean_expression<HIGH_EXPR_MU)]
high_transcribed_starts = starts_by_mean_expression.loc[starts_by_mean_expression>=HIGH_EXPR_MU]
enriched_oligo_starts = starts_by_mean_expression.loc[(starts_by_mean_expression>=MODERATE_EXPR_MU)]

low_transcribed_start_positions =  unique_starts.loc[low_transcribed_starts.index].starts
mid_transcribed_start_positions =  unique_starts.loc[moderately_transcribed_starts.index].starts
high_transcribed_start_positions = unique_starts.loc[high_transcribed_starts.index].starts
enriched_oligo_start_positions = unique_starts.loc[enriched_oligo_starts.index].starts

high_transcribed_start_neighborhoods = pd.concat([ high_transcribed_start_positions + i * 30 for i in range(-2,3)])
mid_transcribed_start_neighborhoods = pd.concat([ mid_transcribed_start_positions + i * 30 for i in range(-2,3)])
enriched_oligo_start_neighborhoods = pd.concat([enriched_oligo_start_positions + i * 30 for i in range(-2,3)])

oligos_by_start = oligos.reset_index().set_index("starts").sort_index()
active_oligos = oligos_by_start.loc[oligos_by_start.groupby(level=0).size() == 5]

mod_oligos = active_oligos.loc[active_oligos.index.isin(mid_transcribed_start_positions)]
high_oligos = active_oligos.loc[active_oligos.index.isin(high_transcribed_start_positions)]
enriched_oligos = active_oligos.loc[active_oligos.index.isin(enriched_oligo_start_positions)]

high_motif_candidates  = high_oligos.loc[high_oligos.groupby(level=0).apply(lambda x: x.mu.loc[x.mu > (x.mu.mean() - x.mu.std())].count() ==4)].groupby(level=0).apply(lambda x:x.sort_values("mu",ascending=True).iloc[0])
mod_motif_candidates  = mod_oligos.loc[mod_oligos.groupby(level=0).apply(lambda x: x.mu.loc[x.mu > (x.mu.mean() - x.mu.std())].count() ==4)].groupby(level=0).apply(lambda x:x.sort_values("mu",ascending=True).iloc[0])
enriched_motif_candidates  = enriched_oligos.loc[enriched_oligos.groupby(level=0).apply(lambda x: x.mu.loc[x.mu > (x.mu.mean() - x.mu.std())].count() ==4)].groupby(level=0).apply(lambda x:x.sort_values("mu",ascending=True).iloc[0])

mod_oligos = mod_oligos.reset_index().set_index("oligo")
high_oligos = high_oligos.reset_index().set_index("oligo")
enriched_oligos = enriched_oligos.reset_index().set_index("oligo")

### active_oligos.reset_index().set_index("oligo") this code had no effect? delete?

#locate enhancer-containing mutants 
oligo_candidates = enriched_oligos
candidate_min_mutant_nums = oligo_candidates.reset_index().groupby("starts").apply(lambda x: x.sort_values("mu").iloc[0].mutant_num)

enhancer_mutant_candidates=  oligo_candidates.loc[
    (oligo_candidates.starts.isin(candidate_min_mutant_nums.index[candidate_min_mutant_nums>0]))
]
candidates_by_start = enhancer_mutant_candidates.reset_index().groupby("starts").apply(lambda x: x.sort_values("mu").iloc[0])
candidates_by_start["wt_sequence"] = oligo_candidates.loc[oligo_candidates.mutant_num==0].set_index("starts").Sequences
enhancer_mut_sequences_by_start = pd.concat([
    candidates_by_start.apply(lambda x: x.wt_sequence[int((x.mutant_num-1) *30 ):int((x.mutant_num-1) *30+30)],axis=1).rename("mutated_sequence"),
    candidates_by_start.wt_sequence,
    candidates_by_start.oligo,
    candidates_by_start.apply(lambda x: x.name,axis = 1).rename("start"),
    (candidates_by_start.apply(lambda x: x.name,axis = 1) + (candidates_by_start.mutant_num-1)*30).rename("mutant_start")  
],axis =1)
enhancer_mut_sequences_by_start.to_csv("../out/0628_motif_candidate_sequences.csv")


#[TODO] follow up on wildtype expression troughs
#note that some of these identified mutants will have the lowest mu for wildtype.
#lets save these and consider separately
wt_silencer_mutant_candidates = oligo_candidates.loc[
    (oligo_candidates.mutant_num== 0) &\
    (oligo_candidates.starts.isin(candidate_min_mutant_nums.index[candidate_min_mutant_nums==0]))]
enhancer_hole_mut_sequences_by_start = wt_silencer_mutant_candidates.reset_index().set_index("starts")


```

### LOAD MOTIFS ###

```python
#load motifs
from Bio.motifs import jaspar as mjaspar
import Bio.motifs
from Bio import motifs as bmotifs
with open("../data/jaspar.pfm") as handle:
    jaspar_motifs = mjaspar._read_jaspar(handle)


#identify motif clusters
from sklearn.cluster import SpectralClustering, MeanShift, KMeans
import numpy as np
jaspar = pd.DataFrame([pd.Series({"consensus":str(j.consensus),"name" :j.name,"pwm":j.pwm,"pssm":j.pssm}) for i,j in enumerate(jaspar_motifs)])
jaspar["spec2_cluster"] = -1
jaspar["ms_cluster"] = -1
jaspar["km_cluster"] = -1
jaspar["ms_center"] = None
jaspar["km_center"] = None

#is this actually helpful?
for l,g in jaspar.groupby(jaspar.consensus.str.len()):
    X = np.array(list(g.pwm.apply(lambda x: np.array([e for v in x.values() for e in v]))))

    #use meanshift clustering to find an ideal n_clusters
    ms_clustering = MeanShift().fit(X)
    jaspar.loc[g.index,"ms_cluster"] =[int(e) for e in ms_clustering.labels_]
    ms_centers = ms_clustering.cluster_centers_
    jaspar.loc[g.index,"ms_center"] = jaspar.loc[g.index]["ms_cluster"].apply(lambda x:ms_centers[x])
    
    #run spectral clustering (not used)
    spec2 = SpectralClustering(
        n_clusters = len(ms_centers),
        assign_labels="discretize",
        random_state=0).fit(X)
    jaspar.loc[g.index,"spec2_cluster"]  =[int(e) for e in spec2.labels_]
    
    #initializing from meanshift, run kmeans clustering to find clusters and canonical motifs
    km_cluster = KMeans(
        n_clusters = len(ms_centers),
        random_state=0).fit(X)
    jaspar.loc[g.index,"km_cluster"] =[int(e) for e in km_cluster.labels_]
    km_centers = km_cluster.cluster_centers_
    jaspar.loc[g.index,"km_center"] = jaspar.loc[g.index]["km_cluster"].apply(lambda x:km_centers[x])

    closest = [ np.argmin([ np.sum(np.square(X[i] -cent))   for i in range(len(X))  ]) for cent in km_centers]
    jaspar.loc[g.index,"km_centroid_jaspar_id"] = jaspar.loc[g.index].km_cluster.apply(lambda i:g.iloc[ closest[int(i)] ].name)

jaspar["len"] = jaspar.consensus.str.len()
jaspar.reset_index().set_index(["len","km_cluster"]).sort_index()
jaspar["cluster_name"] = jaspar.apply(lambda x:f"{x.len}_{x.km_cluster}",axis=1)

```

### ANALYZE MOTIFS ###

```python

wt_oligos = oligos.loc[oligos.mutant_num == 0]
#mut_sequences["max_expressed_wt_oligo"]
wt_starts = wt_oligos.reset_index().set_index("starts")

max_starts_by_sequence = enhancer_mut_sequences_by_start.reset_index().groupby("mutated_sequence")\
    .apply(lambda x: wt_starts.loc[x.start.values].mu.idxmin()) 
max_oligos_by_sequence =max_starts_by_sequence.apply(lambda x: wt_starts.loc[x].oligo)

mseq_groups = enhancer_mut_sequences_by_start.reset_index().groupby("mutated_sequence")
#this oligo has the most difference between wild-type and motif expression
max_delta_mut_oligo =mseq_groups.apply(lambda x: \
                                   (x.set_index("oligo").apply(lambda y:wt_starts.loc[y.start].mu,axis=1) \
                                    - oligos.loc[x.oligo.values].mu).idxmax())
max_delta_wt_oligo = max_delta_mut_oligo.apply(lambda x:wt_starts.loc[oligos.loc[x].starts].oligo)

mutants = enhancer_mut_sequences_by_start.drop_duplicates(subset = ["mutated_sequence"]).set_index("mutated_sequence")[["mutant_start"]]
mutants["seq"] = mutants.index
mutants = mutants.reset_index()
mutants.index.name = "index"
mutants["highest_overlapping_wt_oligo"] = max_oligos_by_sequence
mutants["max_delta_wt_oligo"] = max_delta_wt_oligo
mutants["max_delta_mut_oligo"] = max_delta_mut_oligo
mutants["wt_mu"] = mutants.max_delta_wt_oligo.apply(lambda x:oligos.mu.loc[x])
mutants["mut_mu"] = mutants.max_delta_mut_oligo.apply(lambda x:oligos.mu.loc[x])
mutants["start_index"] = mutants.max_delta_wt_oligo.apply(lambda x:oligos.start_index.loc[x])
pwm_sum_grid = jaspar.pwm.apply(lambda pwm:  
                 mutants.apply(lambda x:
                        np.max(
                [  np.sum([pwm[l][i] for i,l in enumerate(x.name[start:start+len(pwm["G"])])])/ len(pwm["G"]) 
                   for start in range(0,len(x.name) - len(pwm["G"]))
                ]
                               ),axis=1
                )
                       )

from Bio import Seq, Alphabet
mutants = mutants.drop(columns="seqseq")
mutants["seqseq"]= mutants.seq.apply(lambda x:Seq.Seq(x,alphabet=Alphabet.IUPAC.IUPACUnambiguousDNA()))
pwm_prod_grid = jaspar.pssm.apply(lambda  pssm: mutants.seqseq.apply(lambda x: sorted(pssm.search(x),key=lambda x:-1*x[1])))
pwm_scores_grid = pwm_prod_grid.unstack().apply(lambda x: 0 if len(x)==0 else x[0][1]).unstack().T
pwm_indices_grid = pwm_prod_grid.unstack().apply(lambda x: 0 if len(x)==0 else x[0][0]).unstack().T

mutants["best_jaspar_id"] = mutants.apply(lambda x: pwm_scores_grid.T.loc[x.name].idxmax(),axis=1)
mutants["best_jaspar_offset"] = mutants.apply(lambda x: pwm_indices_grid.T.loc[x.name][x.best_jaspar_id],axis = 1)
mutants["best_jaspar_score"] = mutants.apply(lambda x: pwm_scores_grid.T.loc[x.name].max(),axis=1)
mutants["jid"] = mutants["best_jaspar_id"]
mutants["best_jaspar_genome_start"] =  mutants.mutant_start + mutants.best_jaspar_offset + (mutants.best_jaspar_offset <0)*30
mutants["best_jaspar_motif_start"]= mutants["best_jaspar_genome_start"] - mutants["mutant_start"]
mutants["best_jaspar_name"] = mutants.apply(lambda x:jaspar.name.loc[x.best_jaspar_id],axis = 1)
mutants["best_jaspar_genome_len"] = mutants.apply(lambda x:len(jaspar.consensus.loc[x.best_jaspar_id]),axis = 1)
mutants["best_jaspar_motif_consensus"]= mutants.best_jaspar_id.apply(lambda x: jaspar.loc[x].consensus)

mutants = mutants.set_index(mutants.apply(lambda x:x.wt_mu - x.mut_mu ,axis =1).rename("diffval"),append=True)\
    .sort_index(level=1).reset_index().reset_index()\
    .set_index("index").rename({"level_0":"diffrank"},axis="columns")

```
